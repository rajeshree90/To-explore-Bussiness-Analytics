```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import plotly
import seaborn as sns
%matplotlib inline
```

```python

store_data = pd.read_csv("SampleSuperstore.csv")
print("Data imported successfully")

store_data
```

```python
store_data.iloc[10]
```

```python
store_data.describe()
```

```python
store_data.hist(figsize=(12,17),bins=30)
```

```python
store_data.plot()
```

```python
store_data.mean()
```

```python
sns.heatmap(store_data.corr(),annot=True)
```

```python
store_data.corr().plot.bar()
```

```python
sns.countplot(x="Category",data=store_data).set_title("Most sold product line")
```

```python
sns.countplot(x="Sub-Category",data=store_data).set_title("Most sold Sub-Category")
```

```python
plt.xticks(rotation=270)
```

```python
plt.scatter(x=store_data['Sales'],y=store_data['Profit'])
plt.xlabel("Total sales")
plt.ylabel("Profit")
plt.show()
```

```python
sns.set_style('ticks')
sns.countplot(x="Region",data=store_data,palette='hls')
```

```python
sns.countplot(x="Region",hue='Category',data=store_data,palette='Accent')
```
